### Feature Selection - Dropping constant features

In this step we will be removing the features which have constant features which have constant features which are actually not important for solving the problem statement

```python
import  pandas as pd

# Make dataframe of the given data
data = pd.DataFrame({
    "A": [1, 2, 3, 1, 2, 4],
    "B": [4, 5, 6, 7, 8, 9],
    "C": [0, 0, 0, 0, 0, 0],
    "D": [1, 1, 1, 1, 1, 1]
})
```

```python
data.head()
```

### variance Threshold
Feature selector that removes all low- variane features
This feature  selection algorithm look only at the feature (X) not the desired outputs (y) and can thus used for unsupervised leanring

In order to use the variance threshold we use  `from sklearn.feature_selection` import  `varianceThreshold`

```python
from sklearn.feature_selection import VarianceThreshold
var_thres = VarianceThreshold(threshold=0)
var_thres.fit(data)
```

```python
# var_thres.get_support()
# in order to understand the lines we can use the 
data.columns[var_thres.get_support()]
```

```python
constant_columns = [column for column in data.columns if column not in data.columns[var_thres.get_support()]]

print(len(constant_columns))
```

```python
constant_columns
```

```python
data.drop(constant_columns, axis=1, inplace=True)
```
